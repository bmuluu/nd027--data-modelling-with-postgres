```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from analysis_queries import (
    top_10_locations_query, most_frequent_time_of_day_query, 
    users_in_level_query, top_10_songs_query
)
```

```python
connection_string = "postgresql://student:student@127.0.0.1/sparkifydb"
```

## Locations with the most plays

```python
try:
    locations = pd.read_sql_query(top_10_locations_query, connection_string)
    plt.bar('location', 'count', data=locations)
    plt.xticks(rotation = 90)
    plt.show()
except Exception as e:
    print(e)
```

## Frequency of plays at hour of day

```python
try:
    hour_frequency = pd.read_sql_query(most_frequent_time_of_day_query, connection_string)
    plt.title("Frequency of plays at hour of day")
    plt.bar('hour', 'count', data=hour_frequency)
except Exception as e:
    print(e)
```

## Number of users in each level

```python
try:
    level_user_counts = pd.read_sql_query(users_in_level_query, connection_string)
    plt.title("Number of users in each level")
    plt.plot('level', 'count', data=level_user_counts)
except Exception as e:
    print(e)
```

## Top 10 songs played

```python
try:
    top_10_songs = pd.read_sql_query(top_10_songs_query, connection_string)
    plt.title("Top 10 songs played")
    plt.bar('title', 'plays', data=top_10_songs)
    plt.xticks(rotation = 90)
    plt.show()
except Exception as e:
    print(e)
```
